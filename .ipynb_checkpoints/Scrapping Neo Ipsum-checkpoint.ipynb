```python
import os
from urllib.parse import urlsplit, parse_qsl
import csv  
from serpapi import GoogleSearch

q = "engineering"

params = {
    "engine": "google_maps",
    "q": q,
    "type": "search",
    "ll": "@-7.23726,-35.91056,14z",
    "api_key": 'e91e1bf1bb0282b3e5c4b2839868258b2114cef04d88cca28ec4fa04c96f72d6',
}

search = GoogleSearch(params)

print("Local results")

page_number = 1
fieldnames = set()
results_to_write = []

while True:
    results = search.get_dict()

    if not "local_results" in results:
      break

    print(f"\nPage #{page_number}")

    local_results = results.get("local_results", [])
    results_to_write.extend(local_results)

    for result in results.get("local_results", []):
        fieldnames.update(result.keys())

        print(
            f"""
Title: {result.get('title', '')}
Address: {result.get('address', '')}
Rating: {result.get('rating', '')}
Reviews: {result.get('reviews', '')}"""
        )

    if "ads_results" in results:
        print("Ads")

    for result in results.get("ads_results", []):
        print(
            f"""
Title: {result.get('title', '')}
Address: {result.get('address', '')}"""
        )

    pagination = results.get('serpapi_pagination', results.get('pagination', {}))

    if not "next" in pagination:
        break

    # Workaround to update search parameters based on the SerpApi response
    # In the future, it will be done internally in GoogleSearch#pagination
    #
    # Sample query: engine=google_maps&google_domain=google.com&hl=en&ll=%4040.7455096%2C-74.0083012%2C14z&q=Coffee&start=20&type=search
    # dict(parse_qsl(urlsplit("engine=google_maps&google_domain=google.com&hl=en&ll=%4040.7455096%2C-74.0083012%2C14z&q=Coffee&start=20&type=search").query))
    # { "engine": "google_maps", "google_domain": "google.com", "hl": "en", "ll": "%4040.7455096%2C-74.0083012%2C14z", "q": "Coffee", "start": 20, "type": "search" }
    search.params_dict.update(
        dict(parse_qsl(urlsplit(pagination["next"]).query))
    )

    page_number += 1
else:
    print("Done")

with open(f'{q}.csv', 'w', encoding='utf8', newline='') as output_file:
    fc = csv.DictWriter(output_file, fieldnames=fieldnames)
  
    fc.writeheader()
    fc.writerows(results_to_write)
```
